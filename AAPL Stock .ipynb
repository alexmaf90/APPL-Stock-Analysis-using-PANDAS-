```python
fig, ax = plt.subplots(figsize=(14,7))

def plot_candle(df, ax): 
    ax.vlines(x = df.index, ymin= df.Low, ymax= df.High, colors= "k", linewidth=1)

    # Decrease
    red = df.query("Open > Close")
    ax.vlines(x = red.index, ymin= red.Close, ymax= red.Open, colors= "r", linewidth=3)

    # Increase
    green = df.query("Open <= Close")
    ax.vlines(x = green.index, ymin= green.Close, ymax= green.Open, colors= "g", linewidth=3)
    ax.xaxis.set_major_locator(dates.MonthLocator())
    ax.xaxis.set_major_formatter(dates.DateFormatter("%b-%y"))
    ax.xaxis.set_minor_locator(dates.DayLocator())
    return df



(raw
.iloc[:, ::2]
.pipe(fix_columns)
.resample("d")
.agg({"Open": "first", "High": "max", "Low": "min", "Close": "last"})
.loc["jan 2022": "jun 2022"]
.pipe(plot_candle, ax)
)
plt.title("AAPL Candelstick plot Jan 2022 - jun 2022")
```

### Moving averages 

- Moving averages are statistical calculations that help smooth out fluctuations in time series data, allowing for a clearer understanding of underlying trends. 
- They provide a way to identify the overall direction of the data by calculating the average of a specific number of consecutive data points. 
- Moving averages are commonly used in finance, economics, and other fields to analyze and forecast data. 
- They assist in reducing noise, detecting patterns, and generating signals for decision-making. 
- Overall, moving averages are valuable tools for trend analysis and forecasting in time series data.

### Exponential moving averages

- The Exponential Moving Average (EMA) is a statistical calculation that provides a weighted average of time series data, giving more weight to recent values. 
- It is more responsive to recent changes compared to other moving average methods. 
- The EMA is calculated using a smoothing factor and is useful for detecting short-term trends and generating timely signals. - - It is commonly used in finance, trading, and technical analysis for short-term analysis and decision-making.

### Technical Analusis 

**OBV** stands for **On-Balance Volume**, which is a technical analysis indicator used to measure the cumulative buying and selling pressure of a financial instrument over a specific period. The OBV formula is relatively simple and consists of the following 

steps:

- Start with an initial value for the OBV, typically set to zero.

For each trading period:

- If the closing price of the current period is higher than the closing price of the previous period, add the volume of the current period to the OBV.
- If the closing price of the current period is lower than the closing price of the previous period, subtract the volume of the current period from the OBV.
- If the closing price of the current period is equal to the closing price of the previous period, the OBV remains unchanged.

Repeat this process for each trading period in the specified time frame.

```python
def calc_ovb(df): 
    df = df.copy()
    df["OBV"]= 0.0
    
    for i in range (1, len(df)): 
        if df["Close"][i] > df["Close"][i-1]: 
            df["OBV"]= df["OBV"][i-1] + df["Volume"]
            
        elif df["Close"][i] < df["Close"][i-1]: 
            df["OBV"]= df["OBV"][i-1] - df["Volume"]
            
        else: df["OBV"]= df["OBV"][i-1] 
    return df

calc_ovb(aapl)       
```

**The Accumulation/Distribution (A/D)**  indicator is a technical analysis tool used to assess the flow of money into or out of a security or market. 
It provides insight into whether buying or selling pressure is dominant. The A/D indicator is based on the principle that the volume of trading activity reflects the buying and selling pressure behind price movements.

To calculate the A/D indicator, you need the following components:

Closing price: The closing price of the security for a given period (e.g., daily, weekly, or monthly).

High price: The highest price reached during the same period.

Low price: The lowest price reached during the same period.

Volume: The trading volume for the period, which represents the number of shares or contracts traded.

The formula for calculating the A/D indicator is as follows:

A/D = Previous A/D + Current Money Flow

Money Flow = [(Closing Price - Low Price) - (High Price - Closing Price)] / (High Price - Low Price)

Current Money Flow = [(Closing Price - Low Price) - (High Price - Closing Price)] / (High Price - Low Price) * Volume

The resulting A/D values can be plotted on a chart to identify divergences, confirm price trends, or anticipate potential reversals. Positive values indicate accumulation (buying pressure), while negative values indicate distribution (selling pressure).

```python
(aapl
 .assign(mf=((aapl.Close - aapl.Low) - (aapl.High - aapl.Close)) / (aapl.High - aapl.Low),
        cmf = lambda df_: df_.mf * df_.Volume, 
         ad= lambda df_: df_.cmf.cumsum()
            )
)
             
```

```python
def calc_ad(df, close_col = "Close", low_col = "Low", high_col="High", vol_col="Volume"):
    close = df[close_col]
    low = df[low_col]
    high = df[high_col]
    return (aapl
            .assign(mf=((close - low) - (high - close)) / (high - low),
                    cmf = lambda df_: df_.mf * df_[vol_col], # lambda to access mf 
                     ad= lambda df_: df_.cmf.cumsum())       # lamnda to access cmf 
            .ad
        
)
    
(aapl
.assign(ad= calc_ad)
.ad
.plot(figsize=(12,6), color="orange"))   
```

```python
aapl['ad'] = calc_ad(aapl) 

# A/D values
fig, ax1 = plt.subplots(figsize=(12, 6))
color = 'orange'
ax1.plot(aapl['ad'], color=color)
ax1.set_ylabel('A/D Indicator', color=color)
ax1.tick_params(axis='y', labelcolor=color)

# second y-axis for the close price
ax2 = ax1.twinx()
color = 'blue'
ax2.plot(aapl['Close'], color=color)
ax2.set_ylabel('Close Price', color=color)
ax2.tick_params(axis='y', labelcolor=color)
plt.title('Accumulation/Distribution (A/D) Indicator and Close Price')

plt.show()
```

The **Relative Strength Income** (RSI) is a popular technical indicator used to measure the strength and momentum of a price trend. It oscillates between 0 and 100, indicating whether a security is overbought or oversold. The RSI calculation involves the following steps:

Determine the time period: Choose a specific time frame to calculate the RSI. The most commonly used period is 14 days, but it can be adjusted according to your preference or the specific security being analyzed.

Calculate the average gain and average loss: Calculate the average gain and average loss over the chosen time period. The gain is the positive difference between the current and previous closing prices, and the loss is the negative difference.

Start by calculating the price difference (change) for each day:

- If the closing price is higher than the previous day's closing price, the change is positive (gain).
- If the closing price is lower than the previous day's closing price, the change is negative (loss).
- If the closing price is the same as the previous day's closing price, the change is zero.

Calculate the average gain and average loss over the time period:
- Sum all the gains over the chosen time period and divide by the number of periods to calculate the average gain.
- Sum all the losses (absolute values) over the chosen time period and divide by the number of periods to calculate the average loss.

Calculate the relative strength (RS): Divide the average gain by the average loss to obtain the relative strength (RS).

- RS = Average Gain / Average Loss

Calculate the RSI: Use the relative strength (RS) to calculate the RSI value using the following formula:

- RSI = 100 - (100 / (1 + RS))

The resulting RSI value will range between 0 and 100. A higher RSI value indicates (>70) an overbought condition, suggesting a potential reversal or price correction. Conversely, a lower RSI value (<30) indicates an oversold condition, implying a potential buying opportunity.

```python
def avg(df, col, window_size=14):
    results = []
    window = []
    for i , value in enumerate(df[col]):
        window.append(value)
        if i < window_size:
            results.append(np.nan)
            
        elif i == (window_size): #average of the first 14 values 
            window.pop(0) 
            results.append(sum(window)/window_size) 
            
        else: #(previous average *13 + current value)/14
            results.append((results[-1] * (window_size-1)+value)/window_size)
            
    return pd.Series(results, index = df.index)
            
```
